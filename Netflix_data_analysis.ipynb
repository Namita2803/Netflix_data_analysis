```python
import pandas as pd

# Use the correct filename here ðŸ‘‡
df = pd.read_csv('netflix_titles_nov_2019.csv')

# Show first 5 rows
df.head()
```

```python
# Shape of the dataset: (rows, columns)
print("Shape of the dataset:", df.shape)

# Check column names and their data types
print("\nColumn Data Types:\n")
print(df.dtypes)

# Check for missing (null) values in each column
print("\nMissing Values in Each Column:\n")
print(df.isnull().sum())

# Quick overview of the dataset
print("\nDataset Info:\n")
df.info()

```

```python
# Convert 'date_added' column to datetime format
df['date_added'] = pd.to_datetime(df['date_added'])

# Check again
df['date_added'].head()
```

```python
df.isnull().sum()

```

```python
# Keep only rows where 'date_added' is not null
df = df[df['date_added'].notnull()]
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style="whitegrid")

# Plot count of Movies and TV Shows
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='pastel')

plt.title('Count of Movies vs TV Shows on Netflix')
plt.xlabel('Type of Content')
plt.ylabel('Count')
plt.show()
```

```python
 # Fill missing values in 'country' with 'Unknown'
df['country'].fillna('Unknown', inplace=True)

# If multiple countries are listed, take only the first one
df['country'] = df['country'].apply(lambda x: x.split(',')[0])

# Count top 10 countries
top_countries = df['country'].value_counts().head(10)

# Plot it
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Set3')

plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract year from 'date_added'
df['year_added'] = df['date_added'].dt.year

# Count of content added each year (drop nulls)
yearly_data = df['year_added'].value_counts().dropna().sort_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_data.index, y=yearly_data.values, marker='o', color='crimson')

plt.title('Content Added to Netflix Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.show()
```

```python
from collections import Counter

# Drop missing values in 'listed_in'
genre_data = df['listed_in'].dropna()

# Split by comma, flatten into one list
all_genres = []
for item in genre_data:
    genres = item.split(',')
    for genre in genres:
        all_genres.append(genre.strip())

# Count top genres
genre_counts = Counter(all_genres).most_common(10)

# Convert to DataFrame for plotting
genre_df = pd.DataFrame(genre_counts, columns=['Genre', 'Count'])

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=genre_df, x='Count', y='Genre', palette='coolwarm')
plt.title('Top 10 Most Popular Netflix Genres (Overall)')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

```python
# Filter data for Movies and TV Shows
movies_df = df[df['type'] == 'Movie']
tvshows_df = df[df['type'] == 'TV Show']
```

```python
def get_top_genres(dataframe, column='listed_in', top_n=10):
    all_genres = []
    for item in dataframe[column].dropna():
        genres = item.split(',')
        all_genres.extend([genre.strip() for genre in genres])

    genre_count = Counter(all_genres).most_common(top_n)
    return pd.DataFrame(genre_count, columns=['Genre', 'Count'])
```

```python
top_movie_genres = get_top_genres(movies_df)
top_tv_genres = get_top_genres(tvshows_df)
```

```python
# Plotting both
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(data=top_movie_genres, x='Count', y='Genre', palette='Reds', ax=axes[0])
axes[0].set_title('Top Genres in Movies')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Genre')

sns.barplot(data=top_tv_genres, x='Count', y='Genre', palette='Blues', ax=axes[1])
axes[1].set_title('Top Genres in TV Shows')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Filter only Movies
movie_durations = movies_df.copy()

# Keep only the numeric part (e.g., "90 min" â†’ 90)
movie_durations['duration'] = movie_durations['duration'].str.replace(' min', '')
movie_durations = movie_durations[movie_durations['duration'].str.isdigit()]
movie_durations['duration'] = movie_durations['duration'].astype(int)

# Plot distribution
plt.figure(figsize=(10, 5))
sns.histplot(movie_durations['duration'], bins=30, kde=True, color='orange')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
# Make a copy of TV Shows data
tv_durations = tvshows_df.copy()

# Clean the 'duration' column to extract number of seasons
tv_durations['duration'] = tv_durations['duration'].str.replace(' Season', '', regex=False)
tv_durations['duration'] = tv_durations['duration'].str.replace(' Seasons', '', regex=False)

# Filter numeric values
tv_durations = tv_durations[tv_durations['duration'].str.isdigit()]
tv_durations['duration'] = tv_durations['duration'].astype(int)

# Count frequency of seasons
season_counts = tv_durations['duration'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=season_counts.index, y=season_counts.values, palette='Purples')
plt.title('Number of TV Shows by Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Count of TV Shows')
plt.grid(True)
plt.show()
```
